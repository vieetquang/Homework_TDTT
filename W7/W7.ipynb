```python
a = list(map(int, input().split()))
b = int(input())
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        if arr[mid] < target:            
            left = mid + 1
        else:
            right = mid - 1
    
    return -1
print(binary_search(a, 0, len(a)-1, b))
```

```python
def count_occurennces(arr, x):
    tong = 0
    for i in range(len(arr)):
        if arr[i] == x:
            tong += 1
    return tong
a = list(map(int, input().split()))
b = int(input())
print(count_occurennces(a, b))
```

# W7A3

```python
def bubble_sort(arr):
    for i in range(len(arr)):
        swapped = False
        for j in range(0, len(arr) - i -1):
            if arr[j] > arr[j +1]:
                arr[j], arr[j+1]= arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
                arr[j], arr[min_idx] = arr[min_idx], arr[j]
    return arr
a = list(map(int, input().split()))
print(bubble_sort(a))
print(selection_sort(a))
```
