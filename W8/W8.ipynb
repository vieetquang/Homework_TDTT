```python
class Cylinder:
    def __init__(self, bankinh, chieucao):
        self.bankinh, self.chieucao = bankinh, chieucao
    def dientichbemat(self):
        return ((self.bankinh) ** 2 * 3.14) * 2 + 2 * 3.14 * self.chieucao * self.bankinh
    def thetich(self):
        return (self.bankinh) ** 2 * 3.14 * self.chieucao
r, h = map(int, input().split())
c = Cylinder(r, h)
print(f'{c.dientichbemat():.2f}\n{c.thetich():.2f}')
```

```python
def namnhuan(year):
    return year % 400 == 0 or (year % 4 == 0 and year % 100 != 0)
def songay(month, year):
    if month in (1, 3, 5, 7, 8, 10 ,12):
        return 31
    if month in(4, 6, 9, 11):
        return 30
    if month == 2 and namnhuan(month):
        return 29
    if month == 2 and not namnhuan(month):
        return 28
class Date:
    def __init__(self, day, month, year):
        self.day, self.month, self.year = day, month, year
    def checkdate(self):
        if self.year < 0:
            return False
        if not (1 <= self.month <= 12):
            return False
        if not (1<= self.day <= songay(self.month, self.year)):
            return False
        return True
    def nextday(self):
        if self.day < songay(self.month, self.year):
            return f'{self.day + 1}/{self.month}/{self.year}'
        if self.month < 12:
            return f'1/{self.month + 1}/{self.year}'
        return f'1/1/{self.year}'
d, m, y = map(int, input().split('/'))
c = Date(d, m, y)
if not c.checkdate():
    print('INVALID')
else:
    print(c.nextday())
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []
    def add_items(self, ten, gia):
        sp = {"name":ten, "price":gia}
        self.items.append(sp)
    def remove_items(self, ten):
        for i in self.items:
            if i["name"] == ten:
                self.items.remove(i)
                break
    def check_rong(self):
        if not self.items:
            print("Cart rỗng")
        else:
            print("Cart ko rỗng")
    def tong_tien(self):
        tong = 0
        for i in self.items:
            tong += i["price"]
        return tong
    def hien_thi_item(self):
        for i in range(len(self.items)):
            print(f"{i+1}.{self.items[i]['name']}\n")
    def xoa_all(self):
        self.items = []   
cart = ShoppingCart()
while True:
    a = int(input("""
    ---Shopping Cart---
    Chọn 1 chức năng:
    1. Thêm sản phẩm
    2. Xóa sản phẩm
    3. Kiểm tra giỏ rỗng
    4. Tính tổng tiền
    5. Xem sản phẩm
    6. Xóa cart
    7. thoát
    """)) 
    if a == 1:
        ten = input()
        gia = int(input())
        cart.add_items(ten, gia)
    elif a == 2:
        ten = input()
        cart.remove_items(ten)
    elif a == 3:
        cart.check_rong()
    elif a == 4:
        print(cart.tong_tien())
    elif a == 5:
        cart.hien_thi_item()
    elif a == 6:
        cart.xoa_all()
    elif a == 7:
        c = input("CÓ thoát khỏi ctr k?")
        if c == "yes":
            break
```

```python
import math
class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError
        if den < 0:
            den = -den
            num = -num
        self.num, self.den = num, den
    def rut_gon(self):
        common = math.gcd(self.num,self.den)
        self.num //= common
        self.den //= common
    def add(self, other):
        self.num = self.num * other.den + other.num * self.den
        self.den = self.den * other.den
        self.rut_gon()
        return self
    def sub(self, other):
        self.num = self.num * other.den - other.num * self.den
        self.den = self.den * other.den
        self.rut_gon()
        return self
    def mul(self, other):
        self.num = self.num * other.num
        self.den = self.den * other.den
        self.rut_gon()
        return self
    def div(self, other):
        if other.num == 0:
            raise ValueError
        self.num = self.num * other.den
        self.den = self.den * other.num
        self.rut_gon()
        return self
    def __str__(self):
        return f'{self.num}/{self.den}'
a, b, op, c, d = input().split()
frac1 = Fraction(int(a), int(b))
frac2 = Fraction(int(c),int(d))
if op == "+":
    print(frac1.add(frac2))
elif op == "-":
    print((frac1.sub(frac2)))
elif op == "*":
    print((frac1.mul(frac2)))
elif op == "/":
    print((frac1.div(frac2)))
```

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner, self.balance = owner, balance
    def deposit(self, x):
        self.balance += x
        return self
    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x
            return self
ten, tien = input().split()
acc = BankAccount(ten, int(tien))
times = int(input())
for i in range(times):
    with_dep, x = input().split()
    x = int(x)
    if with_dep == "WITHDRAW":
        acc.withdraw(x)
    elif with_dep == "DEPOSIT":
        acc.deposit(x)
print(acc.balance)
```

```python
import math
class Vector:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z
    def tichcohuong(self, other):
        xphaptuyen = self.y * other.z - self.z * other.y
        yphaptuyen = self.z * other.x - self.x * other.z
        zphaptuyen = self.x * other.y - self.y * other.x
        return Vector(xphaptuyen, yphaptuyen, zphaptuyen)
    def dolon(self):
        return math.sqrt((self.x**2) + (self.y**2) + (self.z**2))
    def tichvohuong(self, other):
        return (self.x * other.x) + (self.y * other.y) + (self.z * other.z) 
class Point:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z
    def vecto(self, other):
        xvecto = other.x - self.x
        yvecto = other.y - self.y
        zvecto = other.z - self.z
        return Vector(xvecto, yvecto, zvecto)
points = []
for i in range(4):
    toax, toay, toaz = map(float, input().split())
    points.append(Point(toax, toay, toaz))
AB = points[0].vecto(points[1])
AC = points[0].vecto(points[2])
DB = points[3].vecto(points[1])
DC = points[3].vecto(points[2])
n1 = AB.tichcohuong(AC)
n2 = DB.tichcohuong(DC)
cos = (abs(n1.tichvohuong(n2)))/(n1.dolon() * n2.dolon())
print(round(math.degrees(math.acos(cos)), 2))

```

```python
class ComplexNumber:
    def __init__(self, real, img):
        self.real, self.img = real, img
    def tich(self, other):
        new_real = self.real * other.real - self.img * other.img
        new_img = self.real * other.img + self.img * other.real
        print(f"{new_real} {new_img}")
r1, im1 = map(int, input().split())
r2, im2 = map(int, input().split())
so1 = ComplexNumber(r1, im1)
so2 = ComplexNumber(r2, im2)
so1.tich(so2)
```

```python
class Triangle:
    def __init__(self, x, y, z):
        self.x, self.y, self.z = x, y, z
    def Stacgiam(self):
        if not (self.x + self.y > self.z and self.x + self. z > self.y and self.y + self.z > self.x):
            return -1
        else:
            p = (self.x + self.y + self.z) / 2
            return math.sqrt(p*(p-self.x)*(p-self.y)*(p-self.z)) 
a, b, c = map(float, input().split())
tamgiac = Triangle(a, b, c)
print(round(tamgiac.Stacgiam(), 2))
```
