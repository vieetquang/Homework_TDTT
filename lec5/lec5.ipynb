```python
a = dict(input().split())
b = dict(input().split())
c = {}
for i in a:
    c[i] = int(b.get(i, 0)) + int(a[i])
for j in b:
    c[j] = int(a.get(j, 0)) + int(b[j])
v = dict(sorted(c.items()))
print(v)
```

# W6A10

```python
a = list(map(int, input().split()))
b = int(input())
c = []
for i in range(len(a)):
    for j in range(len(a)):
        if a[i] + a[j] == b and i <= j:
            k = (a[i] , a[j])
            if k not in c:
                c.append(k)
print(c)
```

# W6A12

```python
a = list(map(int, input().split()))
times = {}
bigg = 0
bigkey = None
for i in a:
    times.setdefault(i , 0)
    times[i] += 1
for x, y in times.items():
    if y > bigg:
        bigg = y
        bigkey = x
    elif y == bigg and x < bigkey:
        bigg = y
print(bigkey)
```
