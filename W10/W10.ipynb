```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))
def sapxep(x, y):
    if len(x) == 0:
        return y
    elif len(y) == 0:
        return x
    else:
        if x[0] < y[0]:
            return [x[0]] + sapxep(x[1:], y)
        else:
            return [y[0]] + sapxep(y[1:], x)
print(sapxep(a,b))

```

# W9A2

```python
matrix = []
while True:
    try:
        row = list(map(int, input().split()))
        matrix.append(row)
    except:
        break
k = int(input())
l = matrix[0][0]
r = matrix[-1][-1]
while l < r:
    mid = (l+r)//2
    sothu = 0
    i = len(matrix)
    j = 0
    while i >=0 and j < len(matrix[0]):
        if matrix[i-1][j] <= mid:
            sothu += i
            j += 1
        else:
            i -= 1
    if sothu < k:
        l = mid + 1
    elif sothu >= k:
        r = mid
print(l)

```

```python
a = int(input())
b = []
for i in range(a):
    donga = input()
    dongb = input()
    b.append([donga, dongb])
c = 0
d = 0
def check(i,j,duo):
        if j == len(duo[1]):
            return "YES"
        elif i == len(duo[0]) and j < len(duo[1]):
            return "NO"
        else:
            if duo[0][i].upper() == duo[1][j]:
                return check(i+1, j+1,duo)
            if duo[0][i].upper() != duo[1][j]:
                return check(i+1, 0,duo)
for cap in b:
    print(check(c,d,cap))
```
