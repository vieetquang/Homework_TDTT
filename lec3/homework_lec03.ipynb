# Bài 16

```python
a = float(input())
def lamtronlen(a):
    if a == int(a):
            return int(a)
    else:
        if a > 0:
            return int(a) + 1
        else:
            return int(a)
def lamtronxuong(a):
    if a == int(a):
            return int(a)
    else:
        if a < 0:
            return int(a) - 1
        else:
            return int(a)
def lamtrongannhat(a):
    if a == int(a):
        return int(a)
    if a < 0:
        if a - int(a) < -0.5:
            return int(a) - 1
        else:
            return int(a)
    if a > 0:
        if a - int(a) >= 0.5:
            return int(a) + 1
        else:
            return int(a)
print(lamtronlen(a), lamtronxuong(a), lamtrongannhat(a))
```
