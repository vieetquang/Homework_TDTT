```python
a = "A1.txt"
def dao_so(path):
    with open(path, "r") as f:
        n = f.read().strip()
        re = n[::-1]
    return re
print(dao_so(a))
```

```python
a = "A4_1.txt"
b = "A4_2.txt"
ten = []
ten2 = []
def anagram(p1,p2):
    with open(p1, "r") as f1:
        id1 = f1.read().strip().upper().split()
        for i in id1:
            ten.extend(list(i))
    with open(p2, "r") as f2:
        id2 = f2.read().strip().upper().split()
        for i in id2:
            ten2.extend(list(i))
    if sorted(ten) == sorted(ten2):
        return True
    return False
print(anagram(a,b))
```
