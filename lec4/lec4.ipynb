# W4A10

```python
n = int(input())
def collatz(a):
    buoc = 1
    while a != 1:
        if a % 2 == 0: 
            a //= 2
        else:
            a = a*3+1
        buoc += 1
    return buoc
buocto = 0 
sobe = 0
for i in range(1, n+1):
    step = collatz(i)
    if step > buocto:
        buocto = step
        sobe = i
print(sobe, buocto)
```
