# W4A10

```python
n = int(input())
def collatz(a):
    buoc = 1
    while a != 1:
        if a % 2 == 0: 
            a //= 2
        else:
            a = a*3+1
        buoc += 1
    return buoc
buocto = 0 
sobe = 0
for i in range(1, n+1):
    step = collatz(i)
    if step > buocto:
        buocto = step
        sobe = i
print(sobe, buocto)
```

```python
so_a = int(input())
so_b = int(input())
tong_uoc_a = 0
tong_uoc_b = 0
for i in range(1, so_a):
    if so_a % i == 0:   
        tong_uoc_a += i
for i in range(1, so_b):
    if so_b % i == 0:   
        tong_uoc_b += i
if tong_uoc_a == so_b and tong_uoc_b == so_a:
    print("True")
else:
    print("False")

```
